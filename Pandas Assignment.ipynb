Basic analysis

```python
Temp_mean=df['Temperature (C)'].mean()
Temp_meadian=df['Temperature (C)'].median()
print('Median of tempreture = ',Temp_mean)
print('Median of Tempreture = ',Temp_meadian)
```

```python
min_windspeed=df['Wind Speed (km/h)'].min()
max_windspeed=df['Wind Speed (km/h)'].max()
print('Minimum wind speed = ',min_windspeed)
print('Maximum wind speed = ',max_windspeed)
```

Filtering Data


```python
temp_above_25=df[df['Temperature (C)']>25]
display(temp_above_25)
```

```python
df['percentage_Humidity']=df['Humidity']*100
#display(df['percentage_Humidity'])
high_humidity=df[df['percentage_Humidity']>90]
display(high_humidity)
```

Data Range


```python
start_date=df['Formatted Date'].min()
end_date=df['Formatted Date'].max()
print('The data is from ',start_date,' to ',end_date,' .')
```

Level=Intermediate


Grouping and aggregation


```python
grouped=df.groupby('Precip Type')[['Temperature (C)','Humidity']].mean()
print(grouped)
```

```python
count=df.groupby('Precip Type')[['Temperature (C)','Humidity']].size()
print(count)
```

Correlation Matrix


```python
Correlation_matrix=df[['Temperature (C)','Apparent Temperature (C)','Humidity','Wind Speed (km/h)']].corr()
display(Correlation_matrix)
```

Weather summary analysis

```python
unique_count=df['Summary'].nunique()
print('Number of unique daily summary = ',unique_count)
```

```python
most_common=df['Summary'].value_counts().idxmax()
print('most common daily summary = ',most_common)
```

```python
mean_value=df.groupby('Summary')['Temperature (C)'].mean()
hottest_summary=mean_value.idxmax()
print('Highest average tempreture = ',hottest_summary)
```

Advanced Filtering

```python
Filtered_data=df[(df['Visibility (km)']<10) & (df['Wind Speed (km/h)']>20)]
display('Filtered data = ',Filtered_data)
```

```python
highest_pressure=df.loc[df['Pressure (millibars)'].idxmax()]
display(highest_pressure)
```
